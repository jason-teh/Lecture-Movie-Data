## How sorting works

Python can sort lists using the built-in `sorted()` function.

- `sorted(movies, key=...)` creates a **new sorted list**.
- The `key` tells Python **what value to sort by**.
- `lambda m: m["rating"]` means: “for each movie `m`, use its rating.”
- `reverse=True` sorts from **largest to smallest**.

You are not implementing this yet — just observe what it does.


### Display movies sorted by rating (separate cell)

This cell creates a new list ordered by rating (high → low) and prints each title + rating.

````python
movies_by_rating = sorted(movies, key=lambda m: m["rating"], reverse=True)

for m in movies_by_rating:
    print(f"{m['title']}: {m['rating']}")
````

### Display movies sorted by profits (separate cell)

This cell creates a new list ordered by profits (high → low) and prints each title + profits.

````python
movies_by_profits = sorted(movies, key=lambda m: m["profits"], reverse=True)

for m in movies_by_profits:
    print(f"{m['title']}: ${m['profits']:,}")
````
